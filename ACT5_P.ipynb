# ¿Que es una Regresion Lineal?

La Regresión Lineal es uno de los algoritmos más fundamentales del aprendizaje supervisado. Su objetivo es encontrar la relación matemática entre una o más variables de entrada (X, variables independientes) y una variable de salida continua (y, variable dependiente).

**¿Cómo se implementa en Python?**
La implementación con scikit-learn sigue siempre el mismo flujo de 4 pasos:

# 1.Importación de librerías

```python
import pandas as pd
datos = pd.read_csv("celsius.csv")
```

# 2.Carga y exploración del dataset

```python
datos.info()
```

```python
datos.head()
```

# 3.Visualización de los datos

```python
import seaborn as sb
sb.scatterplot(x='Celsius', y='Fahrenheit', data=datos,
               hue= "Fahrenheit", palette="coolwarm")
```

```python
#Caracteristicas (X), etiqueta(y)

x = datos["Celsius"]
y = datos["Fahrenheit"]
```

```python
y
```

```python
x.values
#[-40, -10, 0, 8...]
#[[-40], [-10], [0], [8],...]
x_pro = x.values.reshape(-1, 1)
y_pro = y.values.reshape(-1, 1)
```

# 4.Predicción de nuevos valores

```python
from sklearn.linear_model import LinearRegression

modelo = LinearRegression()
```

```python
#Entrenamiento

modelo.fit(x_pro, y_pro)
```

```python
celisius = 7900
pre = modelo.predict([[7900]])
print(f"{celisius} grados celsius son {pre} grados fahrenheit")
```

```python
modelo.score(x_pro, y_pro)
```
