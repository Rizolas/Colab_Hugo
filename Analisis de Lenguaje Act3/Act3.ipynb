```python
#Librerias
import requests
from bs4 import BeautifulSoup
import re
```

```python
# 1. Obtenemos el HTML de la página (usaremos una de prueba común)
url = "https://elfondoenlinea.com/"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
print(soup)
```

```python
import pandas as pd

titulo = soup.find_all('p', class_='titulo-mini-slider')
autor = soup.find_all('p', class_='autor-mini-slider')
precios = soup.find_all('ul', class_='nav fce-precios')

ln = [n.text for n in titulo]
ln2 = [n.text for n in autor]
ln3 = [n.text for n in precios]

df = pd.DataFrame({"Titulo": ln, "Autor": ln2, "Precio": ln3,})
df
```
