```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd

url = "https://www.mercadolibre.com.mx/ofertas/novedades-de-temporada"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, 'html.parser')
print(soup)
```

```python
titulo = soup.find_all('h3', class_='poly-component__title-wrapper')
precio = soup.find_all('div', class_='poly-component__price')

ln = [n.text for n in titulo]
ln2 = [n.text for n in precio]

df = pd.DataFrame({"Titulo": ln, "Precio": ln2,})
df
df.to_csv('Mercado Libre.csv', index=False, encoding='utf-8')
```
